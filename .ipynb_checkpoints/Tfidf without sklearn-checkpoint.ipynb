```python
#Imports 
from nltk.tokenize import word_tokenize
import os
import re
import math
import pandas as pd
import numpy as np
```

```python
#Tokenization of each text file
tokenized_files=[]
os.chdir("C:\\Users\\Hp\\Desktop\\Text Mining\\bbcsport")
directories = os.listdir()
for directory in directories:
    path=os.path.join(os.getcwd(),directory)
    os.chdir(path)
    files = os.listdir()
    for file in files:
        with open(file,'r') as curr_file:
            s = curr_file.read()
            s = re.sub(r'[^\w\s]','',s)
            tokenized_files.append(word_tokenize(s))
    os.chdir('..')
```

```python
#Creating a unique set of all words
wordset=[]
for listi in tokenized_files:
    for words in listi:
        if not words.isdigit():
            wordset.append(words)
        
wordset = list(set(wordset))
```

```python
#Calculating the frequency of occurence of a specific word among all documents
wordFreq={}
for word in wordset:
    count=0
    for listj in tokenized_files:
        if word in listj:
            count+=1
    wordFreq[word]=count
```

```python
#Calculating the tf-idf values for all words in the wordset for all documents
directory_tfid={}
total = len(tokenized_files)
file_number=0
for i in tokenized_files:
    file_tfidf = {}
    file_number+=1
    n = len(i)
    temp_list=list(set(i))
    for j in wordset:
        a = i.count(j)
        tf = a/float(n)
        idf= math.log(total/float(wordFreq[j]))
        if tf*idf!=0:
            file_tfidf[j]=tf*idf
    directory_tfid[file_number]=file_tfidf
        
```

```python
num_of_clusters=5
centre1=(np.random.uniform(low=-0.0100,high=0.0100,size=len(wordset))).tolist()
centre2=(np.random.uniform(low=-0.0100,high=0.0100,size=len(wordset))).tolist()
centre3=(np.random.uniform(low=-0.0100,high=0.0100,size=len(wordset))).tolist()
centre4=(np.random.uniform(low=-0.0100,high=0.0100,size=len(wordset))).tolist()
centre5=(np.random.uniform(low=-0.0100,high=0.0100,size=len(wordset))).tolist()
```

```python
#Fucntion to decide cluster using Euclidean distance
def ClusterDecider(temp):
    #scores keep the distance between different centres and the point
    score1 = 0
    score2 = 0
    score3 = 0
    score4 = 0
    score5 = 0
    
    #Euclidean Distance
    for i in range(len(wordset)):
        word = wordset[i]
        #As the list is skewd if word not present value of the co-orinats for the ith dimension is 0 else as calculated
        x = 0
        if word in temp:
            x=temp[word]
        score1+=np.square(centre1[i]-x)
        score2+=np.square(centre2[i]-x)
        score3+=np.square(centre3[i]-x)
        score4+=np.square(centre4[i]-x)
        score5+=np.square(centre5[i]-x)
    
    score1 = np.sqrt(score1)
    score2 = np.sqrt(score2)
    score3 = np.sqrt(score3)
    score4 = np.sqrt(score4)
    score5 = np.sqrt(score5)
    
    temp_list = []
    temp_list.append(score1)
    temp_list.append(score2)
    temp_list.append(score3)
    temp_list.append(score4)
    temp_list.append(score5)
    min_distance = min(temp_list)
    if min_distance == score1:
        return 1
    if min_distance == score2:
        return 2
    if min_distance == score3:
        return 3
    if min_distance == score4:
        return 4
    if min_distance == score5:
        return 5
```

```python
cluster1=[]
cluster2=[]
cluster3=[]
cluster4=[]
cluster5=[]

for i in range(443): #Using 60% of the dataset. Total number of text documents = 737|
    file_number = np.random.randint(1,737)
    temp_file = directory_tfid[file_number]
    cluster = ClusterDecider(temp_file)
    if cluster==1:
        cluster1.append(temp_file)
    elif cluster==2:
        cluster2.append(temp_file)
    elif cluster==3:
        cluster3.append(temp_file)
    elif cluster==4:
        cluster4.append(temp_file)
    elif cluster==5:
        cluster5.append(temp_file)
        
print("Cluster1 :"+str(len(cluster1)))
print("Cluster2 :"+str(len(cluster2)))
print("Cluster3 :"+str(len(cluster3)))
print("Cluster4 :"+str(len(cluster4)))
print("Cluster5 :"+str(len(cluster5)))
```
